# Lineages of Planaria: varying the number of neighbors

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import scanpy.api as sc

sc.settings.verbosity = 2  # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.settings.set_figure_params(dpi=120)  # low dpi (dots per inch) yields small inline figures
sc.logging.print_version_and_date()
sc.logging.print_versions_dependencies_numerics()
```

Some plotting parameters.

```python
aga_plot_params = dict(
    root='neoblast 1',
    layout='rt_circular',
    node_size_scale=0.5,
    node_size_power=0.9,
    max_edge_width=0.7,
    fontsize=3.5)
```

Reading the data.

```python
adata = sc.read('./write/planaria_171202_R.h5ad')
```

## AGA

```python
sc.tl.aga(adata, groups='clusters', n_neighbors=5)
```

```python
sc.pl.aga_graph(adata, title='5 neighbors', **aga_plot_params, save='_5n')
```
