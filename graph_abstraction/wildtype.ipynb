# Planarian lineages: wildtype samples

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import scanpy.api as sc

sc.settings.verbosity = 2  # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.settings.set_figure_params(dpi=150)  # low dpi (dots per inch) yields small inline figures
sc.logging.print_versions()
```

Some plotting parameters.

```python
paga_plot_params = dict(
    legend_fontsize=5,
    root='neoblast 1',
    layout='rt_circular',
    node_size_scale=0.5,
    node_size_power=0.9,
    max_edge_width=0.7,
    fontsize=3.5)
```

Get the data.

```python
adata = sc.read('./write/planaria.h5ad')
```

```python
adata_full = sc.read('./data/171202/reg.dge.txt', cache=True).T
```

Reduce to the 7377 wildtype cells.

```python
adata = adata[adata_full.obs_names.str.startswith(('plan1', 'plan2', 'dmso', 'life'))]
adata.n_obs
```

Compute the tSNE coordinates.

```python
sc.tl.tsne(adata)
```

Compute PAGA graph.

```python
sc.pp.neighbors(adata, n_neighbors=30)
sc.tl.paga(adata, groups='clusters')
```

```python
sc.pl.paga_compare(adata, title='wildtype cells only, tSNE and abstracted graph', solid_edges='confidence_tree', dashed_edges='confidence', **paga_plot_params, save='_wildtype')
```
